```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
# Realizando a conexão ao Banco de Dados
con = create_engine('postgresql+psycopg2://postgres:postgres@localhost:5432/ecommerce')
```

1. Selecione todos produtos de uma categoria.

```python
query1 = """
    SELECT 
        p.nome AS produto,
        p.descricao AS descricao_produto,
        p.preco AS preco,
        p.estoque AS quantidade_estoque,
        c.nome AS categoria,
        c.descricao AS descricao_categoria
    FROM produto p
    INNER JOIN categoria c  
    ON p.id_categoria = c.id_categoria
    WHERE c.id_categoria = 10000
"""

df = pd.read_sql(query1, con)
df
```

2. Encontre o nome e o endereço dos clientes que fizeram um pedido em 2024.

```python
query2 = """
    SELECT 
        c.nome,
        c.endereco,
        p.data_pedido
    FROM cliente c
    INNER JOIN pedido p  
    ON c.id_cliente = p.id_cliente
    WHERE EXTRACT(YEAR FROM p.data_pedido) = 2024
"""

df = pd.read_sql(query2, con)
df
```

3. Liste os produtos cujo preço está entre 50 e 100.

```python
query3 = """
    SELECT 
        nome AS produto,
        descricao,
        preco,
        estoque AS quantidade_estoque 
    FROM produto
    WHERE preco BETWEEN 50 AND 100
"""

df = pd.read_sql(query3, con)
df
```

4. Conte o número total de pedidos feitos no mês de julho de 2024.

```python
query4 = """
    SELECT 
        COUNT(*) AS pedidos_julho_2024
    FROM pedido
    WHERE EXTRACT(MONTH FROM data_pedido) = 07
        AND EXTRACT(YEAR FROM data_pedido) = 2024
"""

df = pd.read_sql(query4, con)
df
```

5. Encontre o total pago por cada pedido.

```python
query5 = """
    SELECT
        ped.id_pedido AS pedido,
        SUM(pag.valor_pago) AS total_pago
    FROM pedido ped
    RIGHT JOIN pagamento pag
    ON ped.id_pedido = pag.id_pedido
    GROUP BY ped.id_pedido
    ORDER BY ped.id_pedido
"""

df = pd.read_sql(query5, con)
df
```

6. Liste o nome dos clientes e o valor total dos seus pedidos.

```python
query6 = """
    SELECT
        c.nome AS clientes,
        p.valor_total AS valor_total_pedidos
    FROM cliente c
    RIGHT JOIN pedido p
    ON c.id_cliente = p.id_cliente
"""

df = pd.read_sql(query6, con)
df
```

7. Encontre os produtos com estoque menor que 10.

```python
query7 = """
    SELECT
        nome AS produto,
        descricao,
        preco,
        estoque
    FROM produto
    WHERE estoque < 10
"""

df = pd.read_sql(query7, con)
df
```

8. Liste os pedidos com seu status e a quantidade de itens em cada pedido.

```python
query8 = """
    SELECT
        p.id_pedido AS pedido,
        p.status,
        SUM(i.quantidade) AS quantidade_de_itens
    FROM pedido p
    INNER JOIN item_do_pedido i
    ON p.id_item_do_pedido = i.id_item_do_pedido
    GROUP BY p.id_pedido, p.status
    ORDER BY p.id_pedido
"""

df = pd.read_sql(query8, con)
df
```

9. Encontre os produtos que não foram vendidos em nenhum pedido.

```python
query9 = """
    SELECT
        pro.nome AS produto,
        pro.descricao,
        pro.preco,
        pro.estoque
    FROM produto pro
    LEFT JOIN item_do_pedido i
        ON pro.id_produto = i.id_produto
    LEFT JOIN pedido ped
        ON i.id_item_do_pedido = ped.id_item_do_pedido
    WHERE ped.id_pedido IS NULL 
"""

df = pd.read_sql(query9, con)
df
```

10. Liste o nome dos 10 maiores clientes e a quantidade de pedidos feitos por cada um (por quantidade de pedidos).

```python
query10 = """
    SELECT 
        c.nome AS cliente,
        COUNT(ped.id_pedido) AS quantidade_pedidos
    FROM cliente c
    INNER JOIN pedido ped 
    ON c.id_cliente = ped.id_cliente
    GROUP BY c.nome
    ORDER BY quantidade_pedidos DESC
    LIMIT 10
"""

df = pd.read_sql(query10, con)
df
```

11. Encontre o produto mais caro e o mais barato.

```python
query11 = """
    SELECT 
        nome AS produto,
        descricao,
        preco AS preco_max_min
    FROM produto
    WHERE preco = (SELECT MAX(preco) FROM produto)
       OR preco = (SELECT MIN(preco) FROM produto)
    ORDER BY preco DESC
"""

df = pd.read_sql(query11, con)
df
```

12. Calcule a idade média dos clientes.

```python
query12 = """
    SELECT 
        AVG(EXTRACT(YEAR FROM AGE(CURRENT_DATE, c.data_nascimento))) AS idade_media_clientes
    FROM cliente c
"""

df = pd.read_sql(query12, con)
df
```

14. Encontre os pedidos feitos por clientes cujo nome começa com a letra 'A'.

```python
query14 = """
    SELECT 
        p.id_pedido AS pedido, 
        p.data_pedido, 
        p.valor_total AS valor,
        p.status,
        c.id_cliente, 
        c.nome AS cliente
    FROM pedido p
    INNER JOIN cliente c
    ON p.id_cliente = c.id_cliente
    WHERE c.nome LIKE 'A%'
    ORDER BY p.id_pedido
"""

df = pd.read_sql(query14, con)
df
```

15. Liste os produtos, sua categoria e uma indicação se estão em estoque ou esgotados (usando CASE).

```python
query15 = """
    SELECT 
        p.id_produto,
        p.nome AS produto,
        c.nome AS categoria,
        CASE 
            WHEN p.estoque > 0 
                THEN 'Em estoque'
                ELSE 'Esgotado'
        END AS status_estoque
    FROM produto p
    INNER JOIN categoria c ON p.id_categoria = c.id_categoria
    ORDER BY p.id_produto;
"""

df = pd.read_sql(query15, con)
df
```
